```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x, a):
    return 1 / (1 + np.exp(-a * x))
```

```python
xx = np.linspace(-5, 5, 100)
```

```python
for a in [0.5, 1, 2, 4]:
    plt.plot(xx, sigmoid(xx, a), label=f'a={a}')
plt.legend()
plt.grid()
```
