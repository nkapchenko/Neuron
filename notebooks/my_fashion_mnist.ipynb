```python
from tensorflow.keras.datasets import fashion_mnist
from neuron.mnist.mnist_loader import nk_vectorized_result, vectorized_result

import neuron.network as network


import numpy as np
from numpy import array
import pandas as pd

import seaborn as sns
from copy import deepcopy

from neuron.activation_functions import Sigmoid
from neuron.target_functions import QuadraticCost, CrossEntropyCost
from neuron.mnist import stepik_original as stepik
from neuron import vizualisation


np.set_printoptions(precision=5)

% load_ext autoreload
% autoreload 2
```

# PARSE DATA

```python
nepochs = 10
batch_size = 10
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

Y_train = np.array([nk_vectorized_result(v) for v in y_train])
Y_test = np.array([nk_vectorized_result(v) for v in y_test])
```

# LEARNING
### Neuron

```python
%%time
# n=5_000 (20 sec) n=12_000(25sec) n=50_000 (2min)
netw = network.Network([784, 30, 10])

netw.fit(X_train, Y_train, epochs=nepochs, batch_size=batch_size, learning_rate = 3., test_data=(X_test, Y_test));

netw.vizualise
```
