```python
import numpy as np
from numpy import array, exp, sqrt
import pandas as pd
from copy import deepcopy
from fox_toolbox.utils import rates, volatility
from neuron.network import Network
from itertools import product
from datetime import datetime, date
from mkt_data import mkt_data
import sklearn
import tensorflow as tf
from sklearn import model_selection
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, Activation, LeakyReLU 

import matplotlib.pyplot as plt
import matplotlib.cm as cm

import seaborn as sns
sns.set_context("poster")
sns.set(rc={'figure.figsize': (16, 9.)})
sns.set_style("whitegrid")

np.set_printoptions(precision=5)

% load_ext autoreload
% autoreload 2
```

```python
n = 100
spots = np.random.uniform(50,150,n)
strikes = spots + np.random.uniform(-10, 10, n)
rates = np.random.uniform(0., 0.1, n)
maturities = np.random.uniform(1.,5., n)
vols = np.random.uniform(0.01, 0.2, n)
```

```python
df = pd.DataFrame([strikes, spots, rates, maturities, vols], index='strike spot rate maturity vol'.split()).T
df['Bachelier'] = df.apply(lambda row: volatility.BachelierPrice(row.spot * np.exp(row.rate * row.maturity), row.strike, row.vol * sqrt(row.maturity), w=1), axis=1)
df.spot /= df.strike
df.Bachelier /= df.strike
df.drop('strike', axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test  = model_selection.train_test_split(df.drop(['Bachelier'], axis=1), df.Bachelier)
```

```python
X_train.head()
```

```python
nodes = 100
model = keras.models.Sequential()

model.add(Dense(nodes, input_dim=X_train.shape[1]))
model.add(LeakyReLU())
model.add(Dropout(0.25))

model.add(Dense(nodes, activation='elu'))
model.add(Dropout(0.25))

model.add(Dense(nodes, activation='relu'))
model.add(Dropout(0.25))

model.add(Dense(nodes, activation='elu'))
model.add(Dropout(0.25))

model.add(Dense(1))
model.add(Activation(tf.nn.softplus))
          
model.compile(loss='mse',optimizer='rmsprop')
```

```python
model.fit(X_train.values, y_train.values, batch_size=64, epochs=30, validation_split=0.1, verbose=2)
```

```python
df_train = deepcopy(X_train)
df_train['Bachelier'] = y_train
df_train['Predict'] = model.predict(X_train.values)

df_test = deepcopy(X_test)
df_test['Bachelier'] = y_test
df_test['Predict'] = model.predict(X_test.values)
```

```python
plt.plot(range(len(df_train)),df_train.Bachelier)
plt.plot(range(len(df_train)),df_train.Predict)
```

```python
model.evaluate(X_train, y_train)
```

```python
plt.plot(range(len(df_test)),df_test.Bachelier)
plt.plot(range(len(df_test)),df_test.Predict)
```
