```python
log = xml_parser.get_files('form_mx')
log2 = xml_parser.get_files('form_A')

dsc, est, _ = xml_parser.get_rate_curves(log)
dsc2, est2, _ = xml_parser.get_rate_curves(log2)

plt.plot(dsc.buckets, dsc.values)
ax = plt.twinx()
ax.plot(dsc2.buckets, dsc2.values, color='r')
```

```python
X = np.linspace(0, 25, 100)
y = dsc(X)
y2 = dsc2(X)

plt.plot(X, y)
ax = plt.twinx()
ax.plot(X,y2, color='r')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), y.reshape(-1, 1), random_state=0)
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X.reshape(-1, 1), y2.reshape(-1, 1), random_state=0)
plt.scatter(X, y2)
plt.scatter(X_test2, y_test2)
plt.ylim((0.014, 0.019))
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)

ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
for reg in [LinearRegression(), Ridge()]:
    reg.fit(X_train, y_train)
    print(reg.__class__.__name__, reg.score(X_test, y_test))
```
